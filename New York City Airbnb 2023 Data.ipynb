```python
#Data cleaning
q1 = df['price'].quantile(0.99)
df = df[df['price'] < q1]

q2 = df['minimum_nights'].quantile(0.99)
df = df[df['minimum_nights'] < q2]

q3 = df['number_of_reviews'].quantile(0.99)
df = df[df['number_of_reviews'] < q3]

q4 = df['reviews_per_month'].quantile(0.99)
df = df[df['reviews_per_month'] < q4]

q5 = df['calculated_host_listings_count'].quantile(0.99)
df = df[df['calculated_host_listings_count'] < q5]
```

```python
grouped = df.groupby(['neighbourhood_group', 'host_id'], as_index=False).size()

# Group by neighborhood and sum the aggregated number_of_listings
total_listings_by_neighbourhood_group = grouped.groupby('neighbourhood_group').sum().reset_index()
total_listings_by_neighbourhood_group.rename(columns={'size': 'count'}, inplace=True)

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=total_listings_by_neighbourhood_group, x='neighbourhood_group', y='count')
```
